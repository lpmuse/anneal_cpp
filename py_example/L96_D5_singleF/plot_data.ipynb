```python
%pylab inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use("Agg")
font = {'size'   : 18}
matplotlib.rc('font', **font)
def plot_action(M=7):
    D = 21
#M = 7
    PATH=300
    B=30
    dirname = 'D%d_M%d/'%(D-1,M)
    minima = np.zeros((PATH*B,2))
    for p in range(200):
        data = np.loadtxt(dirname+'path/D%d_M%d_PATH%d.dat'%(D,M,p))
        minima[B*p:B*(p+1),:] = data[:,[0,2]]
    for p in range(100):
        data = np.loadtxt(dirname+'path0/D%d_M%d_PATH%d.dat'%(D,M,p))
        p=p+200
        minima[B*p:B*(p+1),:] = data[:,[0,2]]
    np.savetxt(dirname+'D%d_M%d_singleF_minima.dat'%(D-1,M), minima, fmt='%e')
    fig = plt.figure(figsize=(12,9))

    ax1 = fig.add_subplot(111)
    ax1.scatter(minima[:,0],minima[:,1])
    ax1.set_title('Lorenz96 SingleF Dim=%d #Meas=%d'%(D-1,M))
    ax1.set_ylabel('Minimized Action')
    ax1.set_xlabel('Log2beta')
    ax1.set_yscale('log')
    plt.savefig('Action_D%d_M%d.png'%(D,M),bbox_inches='tight')
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use("Agg")
font = {'size'   : 18}
matplotlib.rc('font', **font)

def plot_data(M=5,PATH=100):
    D = 21
    B=30
    dirname = 'D%d_M%d/'%(D-1,M)
    minima=np.loadtxt(dirname+'D%d_M%d_singleF_minima.dat'%(D-1,M))
    pset = range(PATH)
    data = np.loadtxt('twin_data.dat')
    NPred = 160
    for p in pset:
        estimate = np.loadtxt(dirname+'path/D%d_M%d_PATH%d.dat'%(D,M,p))
        assim = estimate[-1,3:].reshape((160,D),order="C")
        prediction = np.loadtxt(dirname+'predict/prediction_D%d_M%d_PATH%d.dat'%(D,M,p))
        fig = plt.figure(figsize=(16,20))
        fig.suptitle('SingleF D=%d M=%d'%(D-1,M))
        for i in range(20):
            ax1 = fig.add_subplot(11,2,i+1)
            ax1.plot(data[:(160+NPred),i],'b-',linewidth=2.0,label='data')
            ax1.plot(assim[:,i],'r--',linewidth=2.0,label='estimate')
            ax1.plot(range(160,160+NPred), prediction[:NPred,i],'g--',linewidth=2.0,label='estimate')
            ax1.set_ylabel('x%d'%(i+1))
    #box = ax1.get_position()
        i=D-1
        ax1 = fig.add_subplot(11,2,21)
        ax1.plot(np.ones(160+NPred)*8.17,'b-',linewidth=2.0,label='data')
        ax1.plot(assim[:,i],'r--',linewidth=2.0,label='estimate')
        ax1.plot(range(160,160+NPred), prediction[:NPred,i],'g--',linewidth=2.0,label='estimate')
        ax1.set_ylabel('f')    
        ax1.set_xlabel('time(0.025)')
        maxf = max([np.max(assim[:,i]), 8.17])
        minf = min([np.min(assim[:,i]), 8.17])
        ax1.set_ylim((minf-0.5,maxf+0.5))
        #ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        ax1 = fig.add_subplot(11,2,22)
        ax1.plot(minima[:,0],minima[:,1],'bo')
        ax1.plot(minima[B*p:B*(p+1),0],minima[B*p:B*(p+1),1],'r^')
        ax1.set_ylabel('Minimized Action')
        ax1.set_xlabel('Log2beta')
        ax1.set_yscale('log')
        print 'finish',p
        plt.savefig(dirname+'figure/compare_D%d_M%d_PATH%d.png'%(D,M,p),bbox_inches='tight')
```
